```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from re import X
df = pd.read_csv("train.csv",index_col = "id",parse_dates = ["date"])
y = df.num_sold.copy()
X = df.drop("num_sold",axis = 1,inplace = False)
```

```python
X_test = pd.read_csv("test.csv",index_col = "id",parse_dates = ["date"])
```

```python
X_test.head()
```

```python
X["year"] = X["date"].dt.year
X_test['year'] = X_test['date'].dt.year
X_test=X_test.drop(columns=['date'])
X = X.drop(columns=['date'])
```

```python
for colname in X.columns:
    X[colname], _ = X[colname].factorize()
    X[colname] = X[colname].astype('category')
```

```python
for colname in X_test.columns:
    X_test[colname], _ = X_test[colname].factorize()
    X_test[colname] = X_test[colname].astype('category')
```

```python
mean = y.mean()
y.fillna(mean,inplace = True)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators = 200,random_state = 0)
model.fit(X,y)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = pd.DataFrame(y_pred,columns = ["num_sold"],index = X_test.index)
y_pred.head()
```

```python
y_pred.to_csv("submission.csv")
```
